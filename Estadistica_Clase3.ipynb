````
# continuemos con iris : 
# Eliminemos la variables object (No es cuantitativa)

# Eliminar / remover una columna (axis=1) del dataframe iris usando el metodo drop
# help(iris.drop)
X = iris.drop(iris.columns[-1], axis = 1)

x = X['sepal_length']

x.head()
````

````
Vals , numVals = np.unique(x, return_counts= True)

# Almacenemos esta informacion en un data frame 
SepalLen = pd.DataFrame({ "Valores" : Vals , "NumVals" : numVals})
SepalLen

# moda 
SepalLen[SepalLen.NumVals == np.max(SepalLen.NumVals)]
````

````
# Medidas de tendencia central :

print("""El promedio aritmetico de sepal_length : %.3f \n 
La media de  de sepal_length : %.3f \n
La moda de  de sepal_length : %.3f"""%(x.mean(),np.median(x) , x.mode())
)
````

````
# Percentiles ,Deciles, cuartiles 

Q = np.quantile(x , [0.25,0.5, 0.75])

P12 = np.quantile(x , 0.12)

D5 = np.quantile(x ,  0.5)
````

````
# Medidas de dispersion:

print("""
La desviacion estandar de sepal_length : %.3f \n 
La varianza de  de sepal_length : %.3f \n
El Coeficiente de variacion de  de sepal_length : %.3f %%"""%(np.std(x), x.var(), np.std(x)/x.mean()*100)
)
````

````
# Medidas de forma y concentracion

print("""
La kurtosis de sepal_length : %.3f \n 
El coeficiente de asimetria de  de sepal_length : %.3f \n"""%(x.kurt(), x.skew())
)
````

````
# deseo la imagen mas grande 
plt.figure(figsize= (10,7))

# Histogramas de las 4 variables numericas (float)
n_bins = 20

# Definimos el numero de filas y columnas  que va tener mi grafico : (2,2)
fig, axs = plt.subplots(2,2)

# [0,0] | [0,1]
# -------------
# [1,0] | [1,1]

# Definamos el elemento [0,0] del grafico 
axs[0,0].hist(iris.sepal_length,bins = n_bins, color= 'y')
axs[0,0].set_title("Sepal Length")
 
# Definamos el elemento [0,1] del grafico 
axs[0,1].hist(iris.sepal_width,bins = n_bins)
axs[0,1].set_title("Sepal Width")

# Definamos el elemento [1,0] del grafico 
axs[1,0].hist(iris.petal_length,bins = n_bins)
axs[1,0].set_title("Petal Length")


# Definamos el elemento [1,1] del grafico 
axs[1,1].hist(iris.petal_width,bins = n_bins, color = 'g')
axs[1,1].set_title("Petal Width")

fig.tight_layout(pad = 1.0)

plt.show()
````
